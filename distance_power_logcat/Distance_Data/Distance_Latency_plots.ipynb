```python
import pandas as pd
import matplotlib.pyplot as plt

# Set global font size
plt.rcParams['font.size'] = 25

# List of files for each group
va_files = {
    'RB-Meta Gen2': 'Latency_VA_RBGEN2_distance.xlsx',
    'HeyCyan': 'Latency_VA_Cyan_distance.xlsx',
    'Even' : 'Latency_VA_Even_distance.xlsx'
}

image_ai_files = {
    'HeyCyan': 'Latency_Image AI_Cyan_distance.xlsx',
    'RB-Meta Gen1': 'Latency_Image AI_RBGEN1_distance.xlsx',
    'RB-Meta Gen2': 'Latency_Image AI_RBGEN2_distance.xlsx'
}

# Line styles cycle
line_styles = ['-', '--', ':', '-.', '-', '--']
colors = ['b', 'g', 'r', 'c', 'm', 'y']  # To differentiate further

# Function to plot for a group
def plot_group(files, group_name):
    fig, ax = plt.subplots(figsize=(16, 8))
    i = 0
    for label, file in files.items():
        df = pd.read_excel(file)
        # Assuming columns are 'Distance', 'Perceived Latency', 'Actual Latency'
        ax.plot(df['Distance'], df['Perceived Latency'], 
                linestyle=line_styles[i], color=colors[i], 
                label=f'{label} Perceived', linewidth=4)
        i += 1
        if not (label == 'Even' and group_name == 'VA'):
            ax.plot(df['Distance'], df['Actual Latency'], 
                    linestyle=line_styles[i], color=colors[i], 
                    label=f'{label} Actual', linewidth=4)
            i += 1
    ax.set_xlabel('Distance (m)')
    ax.set_ylabel('Latency (s)')
    ax.legend(fontsize=25, loc='upper left')
    ax.tick_params(axis='both', labelsize=25)
    ax.grid(True, which="both", ls="--")

    plt.tight_layout()
    plt.show()

# Plot for VA
plot_group(va_files, 'VA')

# Plot for Image AI
plot_group(image_ai_files, 'Image AI')
```
