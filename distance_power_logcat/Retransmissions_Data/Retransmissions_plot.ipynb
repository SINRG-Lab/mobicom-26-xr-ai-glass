```python
import pandas as pd
import matplotlib.pyplot as plt

# Set global font size
plt.rcParams.update({'font.size': 35})

# Read the CSV files
df1 = pd.read_csv('T1_Image_cyan_50d_WS_Retransmissions.csv')
df2 = pd.read_csv('T1_VA_RBG1_50d_WS_Retransmissions.csv')
df3 = pd.read_csv('T1_VA_cyan_50d_WS_Retransmissions.csv')
df4 = pd.read_csv('T1_Image_AI_RB_50d_WS_Retransmissions.csv')

# Skip the first 10 seconds (intervals 0-9)
df1 = df1[df1['Interval start'] >= 10]
df2 = df2[df2['Interval start'] >= 10]
df3 = df3[df3['Interval start'] >= 10]


# Restart interval to 0
df1['Interval start'] -= 10
df2['Interval start'] -= 10
df3['Interval start'] -= 10

# Create the plot
fig, ax = plt.subplots(figsize=(20, 8))

# Bar width and offsets for grouped bars
width = 1

# Plot each dataset as bar plots with offsets
ax.bar(df1['Interval start'] - width, df1['Retransmissions'], width=width, label='Cyan + AI Image', align='center')
ax.bar(df2['Interval start'], df2['Retransmissions'], width=width, label='RB-Meta + AI Voice', align='center')
ax.bar(df3['Interval start'] + width, df3['Retransmissions'], width=width, label='Cyan + AI Voice', align='center')
ax.bar(df4['Interval start'] + width, df4['Retransmissions'], width=width, label='RB-Meta + AI Image', align='center')
# Set labels
ax.set_xlabel('Time (s)')
ax.set_ylabel('Packets Retransmitted')
ax.grid(axis='y', linestyle='--', alpha=0.7)
# Add legend
ax.legend(fontsize=25, loc='upper right')

plt.show()
plt.savefig('../Plots/retransmissions.png')
```
